```python
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
    from tensorflow.keras.preprocessing.image import ImageDataGenerator

    # Define the data augmentation parameters
    train_datagen = ImageDataGenerator(
        rescale=1.0/255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True
    )

    # Generate batches of augmented data from the training directory
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(500, 555),
        batch_size=32,
        class_mode='categorical'
    )

    # Set the image dimensions
    image_height = 555
    image_width = 500
    num_classes = 3  # Replace with the actual number of classes in your dataset

    # Set the directories for training and testing
    train_dir = 'C:\\Users\\Juan Siahaan\\Documents\\ABANGKIT23\\Annthea\\train'
    test_dir = 'C:\\Users\\Juan Siahaan\\Documents\\ABANGKIT23\\Annthea\\test'

    # Create the image data generators
    train_datagen = ImageDataGenerator(rescale=1.0/255)
    test_datagen = ImageDataGenerator(rescale=1.0/255)

    # Generate batches of augmented data from the training directory
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(image_height, image_width),
        batch_size=32,
        class_mode='categorical'
    )

    # Generate batches of augmented data from the testing directory
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(image_height, image_width),
        batch_size=32,
        class_mode='categorical'
    )

    # Create the CNN model
    model = Sequential()

    # Add the first convolutional layer
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(image_height, image_width, 3)))
    model.add(MaxPooling2D((2, 2)))

    # Add the second convolutional layer
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Add the third convolutional layer
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Flatten the output and add a dense layer
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))

    # Add the output layer with the appropriate number of classes
    model.add(Dense(num_classes, activation='softmax'))

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(
        train_generator,
        steps_per_epoch=len(train_generator),
        epochs=10,
        validation_data=test_generator,
        validation_steps=len(test_generator)
    )

    from tensorflow import keras

    # Melatih model...

    model.save('C:\\Users\\Juan Siahaan\\Documents\\ABANGKIT23\\Annthea\\model_saved')
```

```python
import tensorflow as tf
import numpy as np
from PIL import Image
import os

# Memuat model yang telah disimpan
model = tf.keras.models.load_model('C:\\Users\\Juan Siahaan\\Documents\\ABANGKIT23\\Annthea\\model_saved\\')

# Mendapatkan daftar file gambar dalam direktori
image_dir = 'C:\\Users\\Juan Siahaan\\Documents\\ABANGKIT23\\Annthea\\Test\\Level0'
image_files = os.listdir(image_dir)

# Menyiapkan list untuk menyimpan hasil prediksi
predictions = []

# Memuat dan menggunakan setiap gambar dalam direktori
for image_file in image_files:
    # Memuat gambar sebagai data masukan
    image_path = os.path.join(image_dir, image_file)
    image = Image.open(image_path)
    image = image.resize((500, 555), resample=Image.BILINEAR)  # Mengubah ukuran gambar dengan mempertahankan aspek rasio
    image_array = np.array(image) / 255.0  # Normalisasi piksel gambar

    # Menambahkan dimensi batch ke dalam array gambar
    input_data = np.expand_dims(image_array, axis=0)

    # Menggunakan model untuk melakukan inferensi
    prediction = model.predict(input_data)

    # Menambahkan hasil prediksi ke dalam list
    predictions.append(prediction)

# Menampilkan hasil prediksi
for image_file, prediction in zip(image_files, predictions):
    print(f"Gambar: {image_file} - Prediksi: {prediction}")
```
